## Conditionals in Python

* if / elif / else

```python
value = 10 

if value > 5:
    print('Inside if block of code') 
    print('Value is greater than 5')

print('Outside if block of code')
```

```python
age = int(input('Enter your age: '))    

if age >= 18:
    print('You are an adult')
else:
    print('You are not an adult')
```

```python
language = input('Enter a programming language: ')

if language.title() == 'Python':
    print('You have selected Python')
elif language.title() == 'C#':
    print('You have selected C#')
elif language.title() == 'Java':
    print('You have selected Java')
else:
    print('You have not selected a valid programming language')
```
