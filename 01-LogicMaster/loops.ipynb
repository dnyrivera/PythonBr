## Loops in Python

* For / While / Do While / Break / Continue/ Pass

```python
count = 0 

while count < 10:
    count += 1
    if count == 3: # jump to the next iteration
        continue
    if count == 7: # break the loop
        break
    print(count)
```

```python
for i in range(10): # range(start, stop, step) 
    if i == 3: # jump to the next iteration
        continue
    if i == 7: # break the loop
        break
    print(i)
```

```python
list = [1, 2, 3, 4, 5]

for num in list:
    print(num)
    
for num in enumerate(list): # enumerate returns a tuple
    print(num)
```

```python
names = ['John', 'Mark', 'Luke', 'Sarah', 'Adam']

for name in names:
    print(name)
```
